```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pumpkins=pd.read_csv('../../data/US-pumpkins.csv')
from sklearn.preprocessing import LabelEncoder
new_columns = ['Color','Origin','Item Size','Variety','City Name','Package']
new_pumpkin=pumpkins.drop([c for c in pumpkins.columns if c not in new_columns],axis='columns')
new_pumpkin
```

```python
new_pumpkin.dropna(inplace=True)
new_pumpkin
```

```python
new_pumpkin = new_pumpkin.apply(LabelEncoder().fit_transform)
new_pumpkin
new_pumpkin.info
```

## Seaborn

```python

g=sns.PairGrid(new_pumpkin)
g.map(sns.scatterplot)
```
